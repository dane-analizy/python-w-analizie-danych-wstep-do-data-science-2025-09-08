## Zadanie

Zrób to co dla gatunków z krajami. Skorzystaj z data frame movies_exploded_df.  
Uwaga - wcześniej wyrzuć wiersze z NaN w kolumnie country.

```python
# trzba odrzucic wiersze z pustym country
movies_exploded_df = movies_exploded_df[movies_exploded_df['country'].notnull()]

# dzielimy string country na listę
movies_exploded_df['county_list'] = movies_exploded_df['country'].apply(lambda s: [el.strip() for el in s.split(",")] )

movies_exploded_df = movies_exploded_df.explode('county_list')

movies_exploded_df
```
