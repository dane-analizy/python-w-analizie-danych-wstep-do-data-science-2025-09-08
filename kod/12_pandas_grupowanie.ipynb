```python
import pandas as pd
import statistics
```

```python
pogoda_df = pd.read_excel("dane/pogoda/big.xlsx")
pogoda_df
```

```python
def pora_dnia(g):
    # 6-10 - rano
    # 11-14 - południe
    # 15-18 - popołudnie
    # 19-22 - wieczór
    # 23-5 - noc

    # if g > 6 and g <= 10:
    if 6 < g <= 10:
        return "rano"
    elif 11 <= g < 14:
        return "południe"
    elif 15 <= g < 18:
        return "popołudnie"
    elif 19 <= g < 21:
        return "wieczór"
    else:
        return "noc"
        
    
        
```

```python
pogoda_df.info()
```

```python
pogoda_df['hour'] = pogoda_df['datetime'].dt.hour
```

```python
pogoda_df
```

```python
pogoda_df['pora_dnia'] = pogoda_df['hour'].map(pora_dnia)
pogoda_df
```

```python
# grupowanie

pogoda_df.groupby('pora_dnia', as_index=False).aggregate(

        {
            "temp": "mean",
            "pressure": "min"
        }
    
)
```

```python
pogoda_df.groupby('pora_dnia', as_index=False).aggregate(
        {
            "temp": ["min", "mean", "max"],
            "pressure": ["min", "max", "mean"],
        }
)
```

```python
pogoda_df.groupby('pora_dnia', as_index=False).aggregate(
        {
            "temp": ["min", "mean", "max"],
            "pressure": ["min", "max", "mean"],
        }
).droplevel(level=1, axis=1)
```

```python
pogoda_agg_df = pogoda_df.groupby('pora_dnia', as_index=False).aggregate(
        {
            "temp": ["min", "mean", "max"],
            "pressure": ["min", "max", "mean"],
        }
)

new_column_names = [f"{k[0]}_{k[1]}" for k in pogoda_agg_df.columns]
pogoda_agg_df = pogoda_agg_df.droplevel(level=1, axis=1)
pogoda_agg_df.columns = new_column_names
pogoda_agg_df
```
