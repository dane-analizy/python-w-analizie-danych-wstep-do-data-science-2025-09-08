```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8], 'C': ['A', 'B']})
```

```python
pd.concat(
    [df1, df2], 
    axis=0
)
```

```python
movies_df = pd.read_csv("dane/film/IMDb_movies.csv",
                       usecols=['imdb_title_id', 'title', 'year', 'country'])

movies_df = movies_df.dropna()
movies_df
```

```python
movie_ratings_df = pd.read_csv("dane/film/IMDb_ratings.csv",
                               usecols=['imdb_title_id', 'total_votes', 'mean_vote']
                              )
movie_ratings_df
```

```python
movies_merged_df = pd.merge(movies_df, movie_ratings_df, how="left", left_on="imdb_title_id", right_on="imdb_title_id")
movies_merged_df
```
