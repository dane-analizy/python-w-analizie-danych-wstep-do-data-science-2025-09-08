```python
data = {'Imie': ['Anna', 'Marcin', None, 'Julia', 'Jan'],
    'Wiek': [28, None, 35, 22, 18],
    'Płeć': ['K', 'M', 'M', 'K', 'M']}
df = pd.DataFrame(data)
df
```

```python
df.dropna()
```

```python
df.dropna(subset=['Imie'])
```

```python
df
```

```python
df.fillna(0)
```

```python
# wypełnienie braków stałą wartością per kolumna
df.fillna({
    'Imie': "brak danych",
    "Wiek": 0
})
```

```python
# wypełnienie braków średnią z całej kolumny
df.fillna({
    'Imie': "brak danych",
    "Wiek": df['Wiek'].mean()
})
```

```python
# wypełnienie braków średnią z grupy w ramach kolumny
df['Wiek_uzupelniony'] = df.groupby('Płeć')['Wiek'].transform( lambda x: x.fillna( x.min() ) )
```
