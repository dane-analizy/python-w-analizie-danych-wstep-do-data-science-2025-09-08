```python
import pandas as pd
```

```python
pogoda_df = pd.read_excel("dane/pogoda/big.xlsx")
pogoda_df
```

```python
temp_q1 = pogoda_df['temp'].quantile(0.25)
temp_q3 = pogoda_df['temp'].quantile(0.75)
temp_iqr = temp_q3 - temp_q1
```

```python
temp_q1 - temp_iqr
```

```python
temp_q3 + temp_iqr
```

```python
pogoda_df
```

```python
pogoda_df.loc[
    (pogoda_df['temp'] < (temp_q1 - temp_iqr) )
    | (pogoda_df['temp'] > (temp_q3 + temp_iqr) )
    ,
    "outlier"
    ] = True

```

```python
pogoda_df
```

```python
pogoda_df.loc[
    (pogoda_df['temp'] >= (temp_q1 - temp_iqr) )
    & (pogoda_df['temp'] <= (temp_q3 + temp_iqr) )
    ,
    'outlier'
    ] = False
    
```

```python
pogoda_df[pogoda_df['temp'] > (temp_q3 + 1.5 * temp_iqr)]
```

```python
pogoda_df
```

```python
pogoda_df.value_counts('outlier')
```

```python
# nowa kolumna z kilku innych
pogoda_df['roznica_temp'] = pogoda_df['temp_max'] - pogoda_df['temp_min']
pogoda_df
```

```python
pogoda_df.value_counts('roznica_temp', normalize=True)
```
