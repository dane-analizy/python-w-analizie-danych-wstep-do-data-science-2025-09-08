```python
import pandas as pd
import seaborn as sns
```

```python
# Dane przykładowe
data = {
    'Region': ['Północ', 'Południe', 'Północ', 'Południe', 'Wschód', 'Zachód'],
    'Produkt': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Kwartal_1': [1200, 1500, 1100, 1300, 1250, 1400],
    'Kwartal_2': [1300, 1600, 1150, 1350, 1280, 1450],
    'Kwartal_3': [1250, 1550, 1125, 1325, 1270, 1425],
    'Kwartal_4': [1350, 1650, 1175, 1375, 1300, 1475],
}

df = pd.DataFrame(data)
df
```

```python
# Region -> Wiersze
# Produkt -> Kolumny
# wartośc z Kw1 =-> wartości


   # values=None,
   #  index=None,
   #  columns=None,
   #  aggfunc: 'AggFuncType' = 'mean',
   #  fill_value=None,


plot_df = df.pivot_table(index='Region', columns='Produkt', values='Kwartal_1', aggfunc='sum', fill_value=0)
plot_df
```

```python
plot_df = df.pivot_table(index='Region', columns='Produkt', values='Kwartal_1', aggfunc='sum')

sns.heatmap(plot_df, cmap="RdYlGn", annot=True, fmt='.0f', )
```
